```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.sample(5)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
# relplot figure level -> squre plot
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter')
```

```python
# use hue parameter
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex')
```

```python
# style parameter
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex', style = 'time')
```

```python
# size parameter
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex', style = 'time', size = 'size')
```

```python
#line plot
sns.lineplot(data = tips, x = 'total_bill', y = 'tip')
```

```python
gap = px.data.gapminder()
```

```python
tem_df = gap[gap['country']=='India']
```

```python
sns.lineplot(data = tem_df, x = 'year', y = 'lifeExp')
```

```python
# using relplot
sns.relplot(data = tem_df, x = 'year', y = 'lifeExp', kind = 'line')
```

```python
# hue -> style
tem_df = gap[gap['country'].isin(['India','Pakistan','China'])]
tem_df
```

```python
# relplot
sns.relplot(data = tem_df, x = 'year', y = 'lifeExp', hue = 'country', kind = 'line')
```

```python
# faceplot
sns.relplot(data = tem_df, x = 'year', y = 'lifeExp', col = 'country', kind = 'line')
```

```python
# both row and column
sns.relplot(data = tem_df, x = 'year', y = 'lifeExp', col = 'country', row = 'continent', kind = 'line')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
# element
sns.displot(data = titanic, x = 'age',kind = 'hist', element = 'step')
```

```python
# add hue parameter
sns.displot(data = titanic, x = 'age',kind = 'hist', hue = 'sex', element = 'step')
```

```python
# faceting using col and row
sns.displot(data = titanic, x = 'age',kind = 'hist', hue = 'sex', element = 'step', col = 'class', row = 'sex')
```

```python
# kdeplot
sns.kdeplot(data = titanic, x = 'age', hue = 'sex',fill=True)
```

```python
# rugplot
sns.displot(data = titanic, x = 'age',kind = 'hist', rug = True)
```

```python
sns.rugplot(data = titanic, x = 'age')
```

#**Matrix plot**

```python
# heatmap
gap
```

```python
gap.pivot(index = 'country', columns = 'year', values = 'lifeExp')
```

```python
temp_df=gap.pivot(index = 'country', columns = 'year', values = 'lifeExp')
sns.heatmap(data = temp_df)
plt.figure(figsize = (15,12))
```

```python
temp_df = gap[gap['continent'] == 'Europe'].pivot(index = 'country', columns = 'year', values = 'lifeExp')
```

```python
sns.heatmap(data = temp_df)
plt.figure(figsize = (15,12))
sns.heatmap(temp_df, annot = True,linewidths=0.5,cmap='viridis')
```

```python
 # cluster map
 sns.clustermap(temp_df, annot = True,linewidths=0.5,cmap='viridis')
```
